```python
%matplotlib inline
from __future__ import division, print_function
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    """Just some convex function"""
    sigma = 10
    return 5*np.exp(np.linalg.norm(x - np.array([-23, 42]))/(2*sigma**2))

def plot_f(f):
    X0 = np.linspace(-100, 100)
    X1 = np.linspace(-100, 100)
    XX0, XX1 = np.meshgrid(X0, X1)
    Xs = np.dstack((XX0, XX1))
    Z = np.apply_along_axis(f, 2, Xs)
    plt.contour(XX0, XX1, Z)
    
plot_f(f)
```

```python
def df(x):
    """Poor man's numerical differentiation of f"""
    eps = 0.00001
    d = np.zeros_like(x)
    for k in range(x.size):
        x1 = x.copy()
        x2 = x.copy()
        x1[k] -= eps
        x2[k] += eps
        d[k] = (f(x2)-f(x1))/(2*eps)
    return d
```

```python
plot_f(f)

xq = 200*np.random.rand(2)-100  # \in R^2
eta = 10
tol = 0.0001

momentum = False
alpha = 0.5
delta_old = 0

for i in range(10000):
    old_xq = xq.copy()
    delta = - eta * df(xq) 
    if momentum:
        delta += alpha * delta_old
        delta_old = delta
    
    xq += delta
    plt.plot([old_xq[0], xq[0]], [old_xq[1], xq[1]], hold=True)
    
    if np.linalg.norm(delta) < tol:
        break

print("Gradient descent stopped after {} iterations with x = {}".format(i, xq))
print("f(x) = {}, df(x) = {}".format(f(xq), df(xq)))
```
